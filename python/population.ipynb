```python
from zipfile import ZipFile
file_names = list() 
with ZipFile('archive.zip', mode='r') as zip:
    zip.extractall()
    for file in zip.infolist():
        file_names.append(file.filename)
        print(file.filename, file.compress_size, file.file_size)
```
